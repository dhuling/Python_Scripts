## Problem 2 ##

Consider the finite discretely sampled time-series $g_{n} = g(n\Delta)$, where $n = -N/2,..0,..N/2$ and $\Delta$ is the sample interval. Find an expression for a **discrete, finite data window** and an expression for a **discrete finite spectral window.**

### 2a. Derive a discrete, finite data window for *$g_{n}$*  ###

*2.1.1 Express the finite discrete time series, $g_{t}^{T}$ in terms of the gate function, a sampling function, and the continuous time series.*  

$g_{t}^{T} = \prod(\frac{t}{N\Delta}) \cdot III(\frac{t}{\Delta}) \cdot g(t)$  
- $\prod(\frac{t}{N\Delta})$ is the gate function (boxcar), centered in time. 


- $III(\frac{t}{\Delta})$ is the sampling function, or shaw function 


- $g_{n} = g(t)$ is the continuous time series



*2.1.2 Express finite, discrete time-series in terms of a sampling data data window, $s_{\Delta}(t)$*  

$g_{t}^{T} = \prod(\frac{t}{N\Delta}) \cdot III(\frac{t}{\Delta}) \cdot g(t) \quad$ <===> $ \quad g_{t}^{T} = s_{\Delta}(t) \cdot g(t)$
- $s_{\Delta}(t) = \prod(\frac{t}{N\Delta}) \cdot III(\frac{t}{\Delta})$



*2.1.3 Derive expression for discrete, finite data window*

$s_{\Delta}(t) = \sum_{n=N/2}^{N/2}w_{n} \delta(t - n\Delta)$
- $w_{n} = \prod(\frac{t}{N\Delta}) = 1$, when $w_{n}$ is a boxcar.

$s_{\Delta}(t) = \sum_{n=N/2}^{N/2} \delta(t - n\Delta)$


### 2b: Derive a discrete, finite sampling window for *$s_{\Delta}(t)$*  ##

*2.2.1 Express $s_{\Delta}(t)$ as $S_{\Delta}(f)$*

$S_{\Delta}(f) = { \int_{-\infty}^{\infty}} \:s_{\Delta}(t) \: e^{-i2\pi f\Delta} dt$

$S_{\Delta}(f) = { \int_{-\infty}^{\infty}} \:\sum_{n = -N/2}^{N/2}\delta(n - \Delta) \: e^{-i2\pi f\Delta} dt$



*2.2.2 From sifting property $S_{\Delta}(f)$ becomes...*

$S_{\Delta}(f) = \sum_{n = -N/2}^{N/2} \: e^{-i2\pi fn\Delta}$



*2.2.3 Multiply left side of equation by 1*

$\Big[S_{\Delta}(f) = \sum_{n = -N/2}^{N/2} \: e^{-i2\pi fn\Delta}\Big]\frac{e^{i \pi f \Delta} - e^{-i \pi f \Delta}}{2i sin(\pi f \Delta)} $
- Where  $1 = \frac{sin(\pi f \Delta)}{sin(\pi f \Delta)} = \frac{e^{i \pi f \Delta} - e^{-i \pi f \Delta}}{2i sin(\pi f \Delta)} $

$ S_{\Delta}(f) = \sum_{n = -N/2}^{N/2} \: \big[e^{-i2\pi fn\Delta}(e^{i \pi f \Delta} - e^{-i \pi f \Delta})\big] (\frac{1}{2i sin(\pi f \Delta)})$


*2.2.4 Use exponent properties to rearrange the two summations*

$\big(\frac{1}{2i sin(\pi f \Delta)}\big) \:\:S_{\Delta}(f)=\sum_{n = -N/2}^{N/2} e^{-i\pi f\Delta(2n-1)} - e^{-i \pi f \Delta(2n+1)}$



$\big(\frac{1}{2i sin(\pi f \Delta)}\big) \:\:S_{\Delta}(f)=\sum_{n = -N/2}^{N/2} e^{-i\pi f\Delta(2n-1)} - \sum_{n = -N/2}^{N/2}e^{-i \pi f \Delta(2n+1)}$


*2.2.5 Show that all terms cancel out besides $n = -N$ and $n = N$. Assume $N = 6$ and $\frac{1}{2i sin(\pi f \Delta)}$ are $-i\pi f\Delta$ are constants *

           k = k+1;

        end
        

*2.2.6 Write $S_{\Delta}(f)$ in terms of $n = -N$ and $n = N$*

$S_{\Delta}(f) = \frac{e^{i\pi fN\Delta} \: - \:e^{-i\pi fN\Delta}}{2i sin(\pi f \Delta)}$


*2.2.7 Write exponentials in terms of cosines and sines*

$S_{\Delta}(f) = \frac{sin(\pi f N \Delta)}{sin(\pi f \Delta)}$
- $sin(\phi) \: = \: \frac{e^{\phi} - e^{-\phi}}{2i}$ is an euler identity

*2.2.7 Multiply by 1 and rewrite ratio in terms of sinc functions*

$\big[S_{\Delta}(f) = \frac{sin(\pi f N \Delta)}{sin(\pi f \Delta)}\big]\frac{\pi f \Delta}{\pi f \Delta}$
